## 3b) Fit a linear model $f(x) = \beta_0 + \beta_1 x$ to the data

- Print the values of the regression coefficients $\beta_0, \beta_1$ determined using least-squares. 
- Plot the fitted model and the scatterplot on the same plot.
- Compute and print the **MSE** and the $R^2$ coefficient for the fitted model.

All numerical outputs should be displayed to three decimal places.

```python
#Your code here
# all_ones = np.ones(np.shape(xvals))
# X_lin = np.column_stack((all_ones, xvals))
X_lin = np.vander(xvals, 2)

beta = np.linalg.lstsq(X_lin, yvals, rcond=None)[0]
print("beta_0 is: ", np.round(beta[0], 2), "beta_1 is: ", np.round(beta[1], 2))
fit = X_lin.dot(beta)

sse_lin = np.linalg.norm(fit - yvals)**2
mse_lin = sse_lin/np.shape(xvals)[0]
print("mse is: ", np.round(mse_lin, 2))

X_0 = np.vander(xvals, 1)

beta_0 = np.linalg.lstsq(X_0, yvals, rcond=None)[0]
fit_0 = X_0.dot(beta_0)

sse_0 = np.linalg.norm(fit_0 - yvals)**2
print("r_2 is: ", np.round((1- (sse_lin/sse_0)), 3))


plt.rcParams['figure.figsize'] = [14, 7]
plt.scatter(xvals, yvals, s=10, color="Black", label="Linear fit")
plt.plot(xvals, fit, color="Red", label="Linear fit")
plt.legend()
plt.show()
```
